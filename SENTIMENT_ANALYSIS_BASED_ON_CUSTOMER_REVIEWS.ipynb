```python
import requests as r
from bs4 import BeautifulSoup as bs

reviews = []

print("IPHONE CHARGER REVIEWS FROM FLIPCART:")
print()

for j in range(1,10):

  url = "https://www.flipkart.com/apple-20w-usb-c-power-charging-adapter-iphone-ipad-airpods/product-reviews/itm692bb2cec57f7?pid=ACCFWSKQJ8GS6DA7&lid=LSTACCFWSKQJ8GS6DA7FQLILC&marketplace=FLIPKART&page={}".format(j)
  source = r.get(url).content

  soup = bs(source,"lxml")

  tag = soup.find_all("div",class_="t-ZTKy")


  for i in tag:

    reviews.append(i.find_all("div")[1].text)

reviews

```

```python
import contractions as c

import re

import nltk

import pandas as pd

from nltk.stem import WordNetLemmatizer

from nltk.corpus import wordnet

wnl = WordNetLemmatizer()

uni = []

data = []

print("AFTER THE LEMMATIZATION:")
for rev in reviews:

  contraction = c.fix(rev)

  punc = '!@#$^&*()~`+={};%:"?/><,.1234567890'

  for i in punc:

    contraction = contraction.replace(i," ")

  rem_emoji = re.compile("["u"\U0001F600-\U0001F64F"
               u"\U0001F300-\U0001F5FF"
               u"\U0001F680-\U0001F6FF"
               u"\U0001F700-\U0001F77F"
               u"\U0001F780-\U0001F7FF"
               u"\U0001F800-\U0001F8FF"
               u"\U0001F900-\U0001F9FF"
               u"\U0001FA00-\U0001FA6F"
               u"\U0001FA70-\U0001FAFF"
               u"\U00002702-\U000027B0"
               u"\U000024C2-\U0001F251""]",flags = re.UNICODE)

  after_rem_emoji = rem_emoji.sub("",contraction).lower()





  tokens = nltk.word_tokenize(after_rem_emoji)

  lemm_tok = [wnl.lemmatize(token,wordnet.VERB) for token in tokens]

  lemmas = " ".join(lemm_tok)


  print(lemmas)

  data.append(lemmas)

  df = pd.DataFrame(data, columns = ["product_rev"])

  uni = uni + lemm_tok

bow = list(set(uni))
```

```python
my_stop_words = ["its","who","ok","i","then",'the','then','for','it','q','we','product…','to','pd','all','bt','fr','on','if','such','you','sho','x','will','they','n','whom','of','iam','that','ll','a','w','be','bis','which','also','at','after','what','want','there','-','sturdy','so','and','my','chargers','an','redo','or','in','up','s','by',]

bag_of_word = []

for w in bow:

 if w not in my_stop_words:

  bag_of_word.append(w)

print('Bag of words:')

print(bag_of_word, '\n')
```

Sentiment analysis is the process of analyzing digital text to determine if the emotional tone of the message is positive, negative, or neutral.


Sentiment analysis typically works by employing natural language processing (NLP) techniques to analyze and understand the sentiment expressed in text. The process involves several steps:


1.  Text Preprocessing
2.  Tokenization

1.  Feature Extraction
2.  Sentiment Classification

1.   Post-processing
2.   Evaluation


Types of Sentiment Analysis
Various types of sentiment analysis can be performed, depending on the specific focus and objective of the analysis. Some common types include:



1.   Document-Level Sentiment Analysis : This type of analysis determines the overall sentiment expressed in a document, such as a review or an article
2.   Sentence-Level Sentiment Analysis : Here, the sentiment of each sentence within a document is analyzed

1.   Aspect-Based Sentiment Analysis : This approach focuses on identifying and extracting the sentiment associated with specific aspects or entities mentioned in the text. For example, in a product review, the sentiment towards different features of the product (e.g., performance, design, usability) can be analyzed separately.

2.   Entity-Level Sentiment Analysis : This type of analysis identifies the sentiment expressed towards specific entities or targets mentioned in the text, such as people, companies, or products

1.   Comparative Sentiment Analysis : This approach involves comparing the sentiment between different entities or aspects mentioned in the text.

















```python
import pandas as pd
from textblob import TextBlob

df = pd.read_csv('charger_rev.csv')

df["sentiment"] = df["product_rev"].apply(lambda x : TextBlob(x).sentiment.polarity)

def pol(polarity):

  if polarity > 0:

    return "positive"

  elif polarity < 0:
    return "negative"

  else:
    return "neutral"

df["sentiment"] = df["sentiment"].apply(pol)

df.to_csv('charger_rev.csv', index=False)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(use_idf = True,smooth_idf = True,ngram_range = (1,1),stop_words = 'english')

tfidf_vec = tfidf.fit_transform(df['product_rev'])

print(tfidf_vec)
```

Several algorithms are commonly used for sentiment analysis, including:
1. Naive Bayes Classifier: Based on Bayes’ theorem, it calculates the probability of a text belonging to a specific sentiment class.
2. Support Vector Machines (SVM): A machine learning algorithm that separates data into different classes using hyperplanes.
3. Recurrent Neural Networks (RNN): Particularly LSTM (Long Short-Term Memory) models, which capture sequential information in text data.
4. Convolutional Neural Networks (CNN): Effective for capturing local patterns in text through convolutional filters.
5. Decision Trees: Constructed based on features of the text to classify sentiments. The choice of algorithm depends on the specific requirements and characteristics of the sentiment analysis task.

```python
from sklearn.model_selection import train_test_split


X_train, X_test, Y_train, Y_test = train_test_split(tfidf_vec, df["sentiment"], test_size=0.25, random_state=10)

from sklearn.naive_bayes import MultinomialNB
MNB = MultinomialNB()
MNB.fit(X_train, Y_train)


from sklearn import metrics
predicted = MNB.predict(X_test)
accuracy_score_1 = metrics.accuracy_score(predicted, Y_test)
print("Accuracy Score1:", accuracy_score_1)
```

```python


from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn import metrics

X_train, X_test, Y_train, Y_test = train_test_split(tfidf_vec, df["sentiment"], test_size=0.25, random_state=42)


svm_classifier = SVC(kernel='linear')

svm_classifier.fit(X_train, Y_train)

predicted = svm_classifier.predict(X_test)

accuracy_score_2 = metrics.accuracy_score(predicted, Y_test)

print("Accuracy Score2:", accuracy_score_2)
```
